## Sets

An unordered collection of **unique** elements. They are mutable (elements can be added/removed) but do not allow duplicate values. Sets are useful for mathematical set operations (union, intersection, difference) and for efficient membership testing.

````python
my_set = {1, 2, 3, 4, 2}
print("Initial set:", my_set)

an_empty_set = set()
my_set.add(5)
my_set.remove(3) # Use .discard() to avoid error if element not present
print("Modified set:", my_set)

# Set operations
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}

print("Union:", set_a.union(set_b))         # Elements in either set
print("Intersection:", set_a.intersection(set_b)) # Elements common to both
print("Difference (A - B):", set_a.difference(set_b)) # Elements in A but not B
print("Is 2 in set_a?", 2 in set_a)          # Membership test
````

## While loops

Both for and while loops are used for iteration, but used for different scenarios:

- for loops: Use a for loop when you know in advance how many times you want to iterate, or when you want to iterate over a sequence (like a list, tuple, string, or range) or any other iterable object. It's ideal when you need to process each item in a collection.

- while loops: Use a while loop when you want to repeat a block of code as long as a certain condition is true. The number of iterations is not known beforehand; the loop continues until the condition becomes false.

````python
target = 10
index = 1
while index <= target:
  print(index, target)
  index += 1
````

## Class

In Python, a **class** is a blueprint for creating **objects** (instances of the class). Objects have attributes (data) and methods (functions that belong to the object).

Key concepts:
*   **`__init__` method**: This is a special method called the constructor. It's automatically called when a new object is created and is used to initialize the object's attributes.
*   **`self` parameter**: Refers to the instance of the class itself. It must be the first parameter of any method in a class, including `__init__`.
*   **Attributes**: Variables that store data for an object. They can be defined inside `__init__` (instance attributes) or directly in the class (class attributes).
*   **Methods**: Functions defined inside a class that perform actions on the object's data.

````python
# Define a simple class named 'Dog'
class Dog:
    # Class attribute (shared by all instances of Dog)
    species = "Canis familiaris"

    # The __init__ method is the constructor
    def __init__(self, name, age):
        # Instance attributes (unique to each Dog object)
        self.name = name
        self.age = age

    # An instance method
    def bark(self):
        return f"{self.name} says Woof!"

    # Another instance method to describe the dog
    def description(self):
        return f"{self.name} is {self.age} years old."

# Create objects (instances) of the Dog class
my_dog = Dog("Buddy", 5)
your_dog = Dog("Lucy", 2)

# Access instance attributes
print(f"My dog's name is {my_dog.name}.")
print(f"Your dog's age is {your_dog.age}.")

# Access class attributes
print(f"All dogs belong to the species: {Dog.species}")
print(f"My dog's species: {my_dog.species}")

# Call instance methods
print(my_dog.bark())
print(your_dog.description())

# You can also change attributes after creation
my_dog.age = 6
print(f"My dog's new age is {my_dog.age}.")
print(my_dog.description())
````

## Python enviroment & package managers


### Main Tools Overview

| Tool | Description |
|------|------------|
| pip | Installs Python packages from PyPI |
| conda | Manages Python + system dependencies |
| mamba / micromamba | Faster versions of conda |
| uv | Fast pip + virtual environment manager |

### pip
- Builds packages locally
- Does NOT manage system dependencies
- installs packages independently
```bash
pip install A B      # fails
pip install A && pip install B   # installs but breaks A
```
### conda
- Manages environments + packages + system dependencies
- Provides precompiled binaries
- Handles CUDA, drivers, system libraries
- Supports non-Python dependencies
- Resolves the entire environment

``` bash
conda create -n myenv python=3.11
conda activate myenv
conda install numpy
```

### pixi (next-gen conda workflow)
``` bash
pixi init
pixi add numpy
pixi run python script.py
```
