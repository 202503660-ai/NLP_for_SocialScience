```python
# 너무 파일이 커서 안됨....

!pip install translate

import pandas as pd
from translate import Translator
from google.colab import files


# 번역기 세팅
translator = Translator(from_lang='ko', to_lang="en")
# 리스트 컴프리헨션 사용
df = pd.Series([translator.translate(x) for x in df_raw['사람문장_통합']])

# 저장
df.to_excel('t_text.xlsx', index=False)
display(df.head())

files.download('t_text.xlsx')
```

```python
# 1. 라이브러리 설치 및 가져오기
!pip install deep-translator pandas tqdm openpyxl

import pandas as pd
from deep_translator import GoogleTranslator
from concurrent.futures import ThreadPoolExecutor
from google.colab import files
from tqdm import tqdm

# 진행률 바 활성화
tqdm.pandas()

# 2. 번역기 세팅
translator = GoogleTranslator(source='ko', target='en')

# 안전한 번역을 위한 함수
def safe_translate(text):
    try:
        if pd.isna(text) or str(text).strip() == "": return text
        return translator.translate(str(text))
    except:
        return text

print("병렬 번역을 시작합니다...")

# ThreadPoolExecutor를 사용해 30개씩 동시에 번역
workers = 30
texts = df_raw['사람문장_통합'].tolist()

with ThreadPoolExecutor(max_workers=workers) as executor:
    # tqdm으로 진행률 표시하며 병렬 처리
    translated_texts = list(tqdm(executor.map(safe_translate, texts), total=len(texts)))

df_raw['영어문장_통합'] = translated_texts

# 저장 및 다운로드
df_raw.to_excel('t_result.xlsx', index=False)
files.download('t_result.xlsx')
```

## Q1: Text cleaning

So far we've learned a few text cleaning perations, let's put them together in a function! This function would be a handy one to refer to if you happen to work with some messy text data, and you want to preprocess it with a single function.

Write a function that has at least the following:
- Lowercase the text
- Remove punctuation marks
- Remove extra whitespace characters

```python
def clean_text(text):

  # 기호 list 생성
  from string import punctuation
  import re
  punc = punctuation

  # Step 1: Lowercase
  text = text.lower()

  # Step 2: Use remove_punct to remove punctuation marks
  # text = re.sub(pattern = r'punc', repl = ' ', string = text)
  text = re.sub(pattern =  f"[{re.escape(punc)}]", repl = '', string = text)

  # Step 3: Remove extra whitespace characters
  text = re.sub(pattern = r'\s+', repl = ' ', string = text)

  # Step 4: More cleaning steps?
  text = text.strip()

  return text
```

```python
# Apply the cleaning function to your text column of the DataFrame
text_column = '영어문장_통합'  # Replace with the actual name of your text column
clean_text_column = 'cleaned_text_eng'  # Name for the new column with cleaned text
df_raw[clean_text_column] = df_raw[text_column].apply(
    lambda x: clean_text(x)
)

display(df_raw[[text_column, clean_text_column]].head())
```

```python
# Your code that removes stopwords from the cleaned text and prints out the results
import pandas as pd
from kiwipiepy import Kiwi
from kiwipiepy.utils import Stopwords

# 1. Kiwi 및 불용어(Stopwords) 객체 초기화
kiwi = Kiwi()
stopwords = Stopwords()

# (선택 사항) 사용자 정의 불용어가 있다면 아래와 같이 추가할 수 있습니다.
# stopwords.add(('제외할단어', 'NNG'))

# 2. 불용어 제거 및 형태소 추출 함수 정의
def remove_stopwords(text):
    # 결측치나 빈 문자열 처리
    if pd.isna(text) or not str(text).strip():
        return ""

    # kiwi.tokenize에 stopwords 객체를 전달하면 자동으로 불용어가 걸러집니다.
    tokens = kiwi.tokenize(str(text), stopwords=stopwords)

    # 걸러진 형태소들의 텍스트(form)만 추출하여 공백으로 연결
    # (리스트 형태로 남기고 싶다면 " ".join()을 빼고 return cleaned_tokens 하시면 됩니다)
    cleaned_tokens = [token.form for token in tokens]
    return " ".join(cleaned_tokens)

# =====================================================================
# [사용자 수정 구간] 본인의 데이터프레임(df)과 컬럼명으로 변경해 주세요.
# =====================================================================

# 예시용 데이터프레임 (이 부분을 지우고 실제 df를 사용하세요)
data = {
    '원본_텍스트': [
        "아니 이것은 정말 유용한 파이썬 코드입니다.",
        "그리고 불용어가 어떻게 제거되는지 한 번 확인해 봅시다."
    ]
}
df = pd.DataFrame(data)

# '원본_텍스트' 자리에 실제 텍스트가 있는 컬럼명을 넣어주세요.
target_column = '원본_텍스트'
new_column = '불용어_제거_텍스트'

# 3. 데이터프레임 컬럼에 함수 적용 (진행 상태를 보려면 apply 대신 tapply 등을 쓸 수도 있습니다)
df[new_column] = df[target_column].apply(remove_stopwords)

# 4. 결과 출력
print(df[[target_column, new_column]])
```

```python
# Compare the results of before and after stopword removal
# ==========================================
# 불용어 제거 전/후 결과 비교 코드
# ==========================================

# 1. 눈으로 나란히 비교하기 (상위 5개 행만 깔끔하게 출력)
print("=== 1. 전/후 텍스트 직접 비교 ===")
display(df[[target_column, new_column]].head()) # 주피터 환경이면 display, 터미널이면 print 사용
print("-" * 50)


# 2. 텍스트 길이(단어 개수) 변화 비교하기
# 띄어쓰기 기준으로 단어가 몇 개인지 세어줍니다.
df['기존_단어수'] = df[target_column].apply(lambda x: len(str(x).split()))
df['제거후_단어수'] = df[new_column].apply(lambda x: len(str(x).split()))
df['감소한_단어수'] = df['기존_단어수'] - df['제거후_단어수']

print("\n=== 2. 단어 수 변화 수치 확인 ===")
display(df[[target_column, '기존_단어수', '제거후_단어수', '감소한_단어수']].head())
print("-" * 50)


# 3. (심화) 실제로 지워진 불용어가 무엇인지 추출하기
def get_removed_words(row):
    # 기존 문장의 띄어쓰기된 단어들을 집합(Set)으로 만듭니다.
    original_words = set(str(row[target_column]).split())

    # 불용어 제거 후 남은 단어들을 집합으로 만듭니다.
    # (형태소 분석이 되면서 모양이 조금 바뀔 수 있어 완벽하게 일치하지 않을 순 있지만 대략적 파악에 매우 좋습니다)
    cleaned_words = set(str(row[new_column]).split())

    # 기존 단어 집합에서 남은 단어 집합을 빼서(차집합) 지워진 단어만 남깁니다.
    removed = original_words - cleaned_words
    return ", ".join(removed)

df['지워진_단어들'] = df.apply(get_removed_words, axis=1)

print("\n=== 3. 제거된 불용어 직접 확인 ===")
display(df[[target_column, '지워진_단어들']].head())
```
