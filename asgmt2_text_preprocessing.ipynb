```python
# 1. 라이브러리 설치 및 가져오기
!pip install deep-translator pandas tqdm openpyxl

import pandas as pd
from deep_translator import GoogleTranslator
from concurrent.futures import ThreadPoolExecutor
from google.colab import files
from tqdm import tqdm

# 진행률 바 활성화
tqdm.pandas()

# 2. 번역기 세팅
translator = GoogleTranslator(source='ko', target='en')

# 안전한 번역을 위한 함수
def safe_translate(text):
    try:
        if pd.isna(text) or str(text).strip() == "": return text
        return translator.translate(str(text))
    except:
        return text

print("병렬 번역을 시작합니다...")

# ThreadPoolExecutor를 사용해 30개씩 동시에 번역
workers = 30
texts = df_raw['사람문장_통합'].tolist()

with ThreadPoolExecutor(max_workers=workers) as executor:
    # tqdm으로 진행률 표시하며 병렬 처리
    translated_texts = list(tqdm(executor.map(safe_translate, texts), total=len(texts)))

df_raw['영어문장_통합'] = translated_texts

# 저장 및 다운로드
df_raw.to_excel('t_result.xlsx', index=False)
files.download('t_result.xlsx')
```

## Q1: Text cleaning

So far we've learned a few text cleaning perations, let's put them together in a function! This function would be a handy one to refer to if you happen to work with some messy text data, and you want to preprocess it with a single function.

Write a function that has at least the following:
- Lowercase the text
- Remove punctuation marks
- Remove extra whitespace characters

```python
def clean_text(text):

  # 기호 list 생성
  from string import punctuation
  import re
  punc = punctuation

  # Step 1: Lowercase
  text = text.lower()

  # Step 2: Use remove_punct to remove punctuation marks
  # text = re.sub(pattern = r'punc', repl = ' ', string = text)
  text = re.sub(pattern =  f"[{re.escape(punc)}]", repl = '', string = text)

  # Step 3: Remove extra whitespace characters
  text = re.sub(pattern = r'\s+', repl = ' ', string = text)

  # Step 4: More cleaning steps?
  text = text.strip()

  return text
```

```python
# Apply the cleaning function to your text column of the DataFrame
text_column = '영어문장_통합'  # Replace with the actual name of your text column
clean_text_column = 'cleaned_text_eng'  # Name for the new column with cleaned text
df_raw[clean_text_column] = df_raw[text_column].apply(
    lambda x: clean_text(x)
)

display(df_raw[[text_column, clean_text_column]].head())
```

```python
import pandas as pd
from kiwipiepy import Kiwi
from kiwipiepy.utils import Stopwords

# 1. Kiwi 및 불용어 객체 초기화
kiwi = Kiwi()
stopwords = Stopwords()

# 2. 불용어 제거 및 형태소 추출 함수 정의
def remove_stopwords(text):
    # 결측치나 빈 문자열 처리
    if pd.isna(text) or not str(text).strip():
        return ""

    # kiwi.tokenize에 stopwords 객체를 전달하면 자동으로 불용어가 걸러집니다.
    tokens = kiwi.tokenize(str(text), stopwords=stopwords)

    # 걸러진 형태소들의 텍스트만 추출하여 공백으로 연결
    cleaned_tokens = [token.form for token in tokens]
    return " ".join(cleaned_tokens)


# 3. 데이터프레임 컬럼에 함수 적용
df_raw['불용어_제거'] = df_raw['cleaned_text_ko'].apply(remove_stopwords)

# 4. 결과 출력
print(df_raw[['cleaned_text_ko', '불용어_제거']])
```

```python
import pandas as pd
from kiwipiepy.utils import Stopwords

# 1. 불용어 객체 불러오기
stopwords = Stopwords()

# 2. 불용어 리스트 꺼내기
if hasattr(stopwords, 'stopwords'):
    if isinstance(stopwords.stopwords, dict):
        stopword_list = list(stopwords.stopwords.keys())
    else:
        stopword_list = list(stopwords.stopwords)
else:
    stopword_list = list(stopwords._stopwords)

# 3. 데이터프레임으로 변환
df_stopwords = pd.DataFrame(stopword_list, columns=['형태소', '품사태그'])

# 4. 전체 갯수 및 결과 확인
print(f" Kiwi 불용어 총 갯수: {len(df_stopwords)}개\n")
display(df_stopwords.head(15), df_stopwords.tail(15), df_stopwords)
```

```python
# Compare the results of before and after stopword removal

# 1. 상위 5개 행만 비교
print("=== 1. 전/후 텍스트 직접 비교 ===")
display(df_raw[['cleaned_text_ko', '불용어_제거']].head()) # 주피터 환경이면 display, 터미널이면 print 사용
print("-" * 50)

# 2. 텍스트 길이(형태소의 갯수) 변화 비교
def count_original_tokens(text):
    if pd.isna(text) or not str(text).strip():
        return 0
    return len(kiwi.tokenize(str(text))) # 기존 문장의 형태소 갯수

def count_cleaned_tokens(text):
    if pd.isna(text) or not str(text).strip():
        return 0
    return len(str(text).split()) # 불용어 제거 후 남은 갯수

df_raw['기존_형태소수'] = df_raw['cleaned_text_ko'].apply(count_original_tokens)
df_raw['제거후_형태소수'] = df_raw['불용어_제거'].apply(count_cleaned_tokens)
df_raw['감소한_갯수'] = df_raw['기존_형태소수'] - df_raw['제거후_형태소수']

print("\n=== 2. 단어(형태소) 수 변화 수치 확인 ===")
display(df_raw[['cleaned_text_ko', '기존_형태소수', '제거후_형태소수', '감소한_갯수']].head())
print("-" * 50)


# 3. 실제로 지워진 불용어 추출
def get_real_removed_words(text):
    if pd.isna(text) or not str(text).strip():
        return ""

    # 1. 불용어 제거 안 한 전체 형태소들
    all_tokens = [t.form for t in kiwi.tokenize(str(text))]
    # 2. 불용어 제거 한 후의 형태소들 (불용어_제거 컬럼과 동일)
    clean_tokens = [t.form for t in kiwi.tokenize(str(text), stopwords=stopwords)]

    # 전체 형태소에서 살아남은 형태소들을 제외한 제거된 불용어들
    removed = [word for word in all_tokens if word not in clean_tokens]

    # 중복 제거 후 쉼표로 연결
    return ", ".join(set(removed))

df_raw['지워진_불용어'] = df_raw['cleaned_text_ko'].apply(get_real_removed_words)

print("\n=== 3. 진짜로 제거된 불용어 확인 ===")
display(df_raw[['cleaned_text_ko', '지워진_불용어']].head())
```

```python
import pandas as pd
from tqdm import tqdm

# Fix: Upgrade sympy to resolve compatibility issues with torch
!pip install --upgrade sympy

from kiwipiepy import Kiwi
from konlpy.tag import Okt, Komoran, Kkma
from transformers import AutoTokenizer

# 진행 상태바(tqdm)를 pandas의 apply와 연동
tqdm.pandas()

print("1. 토크나이저 및 모델 초기화 중 (시간이 조금 걸릴 수 있습니다)...")
kiwi = Kiwi()
okt = Okt()
komoran = Komoran()
kkma = Kkma()
# KLUE-BERT 토크나이저 로드 (Hugging Face)
bert_tokenizer = AutoTokenizer.from_pretrained("klue/bert-base")

# 2. 안전한 토크나이징 함수 정의
def get_kiwi_tokens(text):
    if not isinstance(text, str) or not str(text).strip(): return []
    return [token.form for token in kiwi.tokenize(text)]

def get_okt_tokens(text):
    if not isinstance(text, str) or not str(text).strip(): return []
    return okt.morphs(text, norm=True, stem=True)

def get_komoran_tokens(text):
    if not isinstance(text, str) or not str(text).strip(): return []
    try: return komoran.morphs(text)
    except: return []

def get_kkma_tokens(text):
    if not isinstance(text, str) or not str(text).strip(): return []
    try: return kkma.morphs(text)
    except: return []

def get_bert_tokens(text):
    if not isinstance(text, str) or not str(text).strip(): return []
    # BERT는 .tokenize()를 사용해 서브워드(Subword) 단위로 분리합니다.
    return bert_tokenizer.tokenize(text)


print("\n2. 데이터프레임 토크나이징 적용 시작...")
target_columns = ['사람문장_통합', 'cleaned_text_ko']

for col in target_columns:
    # 만약 해당 컬럼이 데이터프레임에 없다면 건너뜀니다.
    if col not in df_raw.columns:
        continue

    print(f"\n[{col}] 컬럼 처리 중...")

    print("  - Kiwi 적용 중...")
    df_raw[f'{col}_kiwi'] = df_raw[col].progress_apply(get_kiwi_tokens)

    print("  - Okt 적용 중...")
    df_raw[f'{col}_okt'] = df_raw[col].progress_apply(get_okt_tokens)

    print("  - Komoran 적용 중...")
    df_raw[f'{col}_komoran'] = df_raw[col].progress_apply(get_komoran_tokens)

    print("  - KLUE-BERT 적용 중...")
    df_raw[f'{col}_bert'] = df_raw[col].progress_apply(get_bert_tokens)

print("토크나이징 완료")
```
